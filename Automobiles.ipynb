```python
import pandas as pd
df=pd.read_csv("imports-85.data", header=  None)
df.head()
```

```python
# to make the first row as the header of the dataset
Header= df.iloc[0]
print(Header)
df = df[1:]
df = df.rename(columns = Header)
df.head()
```

```python
df.dtypes
```

```python
#remove data/ row  that doesn't have "price" mentioned
df.dropna( subset =["price"] , axis=0)
```
